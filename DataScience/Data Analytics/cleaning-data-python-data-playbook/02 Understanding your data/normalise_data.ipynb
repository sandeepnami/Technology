```python
import pandas as pd
```

```python
data = pd.read_csv("artwork_sample.csv")
data.head()
```

```python
data.dtypes
```

```python
print(data.height.min())
print(data.height.max())
print(data.height.mean())
print(data.height.std())
```

```python
height = data.height
```

```python
norm = (height - height.mean())/height.std()
norm
```

```python
minmax = (height - height.min())/(height.max()-height.min())
minmax
```

```python
minmax.min()
```

```python
minmax.max()
```

```python
data
```

```python
data.height = minmax
```

```python
data['standardised_height'] = norm
```
