# Normalizing Data

```python
data.height
```

```python
height_series = data.height
norm = (height_series - height_series.mean())/height_series.std()
norm
```

```python
minmax_series = (height_series - height_series.min())/(height_series.max() - height_series.min())
minmax_series
```

```python
minmax_series.min()
```

```python
minmax_series.max()
```

```python
data['standardised_height'] = norm
```
