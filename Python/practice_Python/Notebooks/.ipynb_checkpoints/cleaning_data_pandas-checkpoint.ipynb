# Normalize Data: Applicable to numeric only
Normalising data means adjusting values of a data and change its scale 

```python
data.head()
```

```python
data.dtypes
```

```python
data.height.mean()
```

```python
# namiobs Mean doesnot work for non numberic data
data.artist.mean()
```

```python
height = data.height
norm = (height - height.mean())/height.std()
norm
```

```python
minmax = (height - height.min())/(height.max() - height.min())
minmax
```

```python
minmax.max()
```

```python
# add new column to data frame
data['standardized_height'] = norm
data.head()
```

```python
# namiobs we cannot add new column to csv by this
data.new_column = norm
```

# Filtering data

```python
data = pd.read_csv('./data/artwork_data.csv', low_memory=False)
```

```python
data.head()
```

```python
data.filter(['id','artist'])
```

```python
data.filter(like='year') # here like is case sensitive
```

```python
data.filter(like='artist')
```

```python
data.filter(regex='(?i)year') # filter using regular expressions -- here (?i): case insensitive
```

```python
# filter on rows
data.filter(regex='^100.$', axis=0) # ^ indicates=> start from 100=> from 1000 row .=>get all rows $=>set range to next 10 rows
```

```python
data.filter(regex='^100(0|2|4|6|8)$', axis=0)
```
