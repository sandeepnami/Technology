## Multi input Comprehensions

```python
[(x,y) for x in range(5) for y in range(5)]
```

```python
# The for loop equivalent of above comprehension
points = []                   
for x in range(5):            
    for y in range(5):        
        points.append((x, y)) 
points
```

```python
# writing comprehensions in multiple lines for readability
values = [x / (x - y) for x in range(100) if x > 50 for y in range(100) if x - y != 0] 

values = [x / (x - y)        
          for x in range(100)
          if x > 50          
          for y in range(100)
          if x - y != 0]     
print(values)
```

```python
# for loop equivalent of above
for x in range(100):                      
    if x > 50:                            
        for y in range(100):              
            if x - y != 0:                
                values.append(x / (x - y))
values
```

```python
[(x, y) for x in range(10) for y in range(x)] 
```

```python
# for loop equivalent of above
result=[]
for x in range(10):          
    for y in range(x):       
        result.append((x, y))
result                             
```

## Nested Comprehensions

```python
vals = [[y * 3 for y in range(x)] for x in range(10)] 
print('nested comprehension=',vals)

# for equivalent of above
outer = []                   
for x in range(10):          
    inner = []               
    for y in range(x):       
        inner.append(y * 3)  
    outer.append(inner)      
print('for loop equivalent of nested comprehension=',outer)
```

```python
[x*y for x in range(10) for y in range(10)] #like multiplication tables data
```

```python
{x*y for x in range(10) for y in range(10)} #note that sets eliminate the duplicates
```
